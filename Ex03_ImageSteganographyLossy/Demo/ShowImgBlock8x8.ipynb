```python
from bitarray import bitarray
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
cover_pixels = np.array(Image.open('lena.bmp'))
```

```python
plt.figure(figsize=(20,20))
#plt.imshow(cover_pixels, cmap='gray')
grid_cover_pixels = cover_pixels.copy()
# Tạo lưới 8x8
grid_cover_pixels[::8, :] = 128 # ::8 Lấy từ đầu đến cuối - bước nhảy là 8 | : là lấy hết
grid_cover_pixels[:, ::8] = 128 # ::8 Lấy từ đầu đến cuối - bước nhảy là 8 | : là lấy hết
# Đánh chỉ số
plt.xticks(range(0, cover_pixels.shape[1], 8))
plt.yticks(range(0, cover_pixels.shape[0], 8))
# Hiển thị ảnh
plt.imshow(grid_cover_pixels, cmap='gray')
```

```python
# Truy xuất block ảnh 8x8 tại vị trí (0, 0)
pos = (0, 0)
cover_pixels[pos[0]:pos[0]+8, pos[1]:pos[1]+8]  # Truy xuất một block 8x8
plt.imshow(cover_pixels[pos[0]:pos[0]+8, pos[1]:pos[1]+8], cmap='gray')
```

=> Ảnh tần số cao  
Mắt người không nhạy cảm với ảnh tần số cao nên nhìn như chỉ có 1 màu
