```python
from bitarray import bitarray
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
cover_pixels = np.array(Image.open('lena.bmp'))
```

```python
plt.figure(figsize=(20,20))
#plt.imshow(cover_pixels, cmap='gray')
grid_cover_pixels = cover_pixels.copy()
# To li 8x8
grid_cover_pixels[::8, :] = 128 # ::8 Ly t u n cui - bc nhy l 8 | : l ly ht
grid_cover_pixels[:, ::8] = 128 # ::8 Ly t u n cui - bc nhy l 8 | : l ly ht
# nh ch s
plt.xticks(range(0, cover_pixels.shape[1], 8))
plt.yticks(range(0, cover_pixels.shape[0], 8))
# Hin th nh
plt.imshow(grid_cover_pixels, cmap='gray')
```
