```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

### Đọc file âm thanh

```python
rate, samples = wavfile.read('brasso.wav')
```

```python
rate
```

```python
samples
```

### Echo - Tiếng vang

#### Độ trễ

```python
time_delay = 1.0 # Thời gian trễ là 1 (tức là giữ nguyên)
sample_delay = int(time_delay * rate)
sample_delay
```

#### Tạo Echo

```python
echo = np.zeros_like(samples[:, 0])   # Tạo mảng toàn phần tử 0
echo[sample_delay:] = samples[:-sample_delay,  0] 
```

```python
# Đồ thị biểu diễn 1 sóng và echo của nó
plt.plot(samples[:, 0])
plt.plot(echo)
```

```python
# Đồ thị biểu diễn ???
samples_echo = samples[:, 0] + echo 
plt.plot(samples_echo)
```

### Ghi file có echo

```python
wavfile.write('brasso_echo.wav', rate, samples_echo)
```
