```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
imgLena = Image.open("lena.gif")
print(imgLena)
print(type(imgLena))
```

```python
# Lấy ma trận mảng 2 chiều - Ma trận chỉ số của bảng màu
pixels = np.array(imgLena)
print(type(pixels))
print(pixels[0])
```

```python
# Lấy bảng màu của ảnh palette dạng list
lstTblColor = imgLena.getpalette()
print(type(lstTblColor))
print("Kích thước: " + str(len(lstTblColor)))  # = 768 = 256 * 3   (256 là số bit của 1 bảng màu | 3 là số màu của ảnh RGB)
print(lstTblColor[:10])
```

```python
# Chuyển bảng màu dạng list thành dạng ma trận với mỗi dòng có 3 giá trị là index của 3 màu RGB
palette = imgLena.getpalette()
nparrayImg = np.array(palette)
#print(nparrayImg)
print(type(nparrayImg))
npReshape = nparrayImg.reshape(1,256,3)   # Tham số đầu = 1: Số dòng
                                          # Tham số đầu hai = -1 hoặc 256: Số cột  (Bằng 1 thì khi đó dựa vào 2 tham số còn lại để tính)
                                          # Tham số ba = 3: Số phần tử của mỗi dòng
# print(len(npReshape))
# print(len(npReshape[0]))
# print(npReshape[0])
# print(len(npReshape[0][0]))
# print(npReshape[0][:5])
```

```python
# Tạo ảnh RGB hiển thị lên từ bảng màu và dãy chỉ số
plt.imshow(npReshape, aspect = 30)    # Điều chỉnh aspect để tăng chiều rộng
```

```python
# Truy xuất giá trị pixels dựa trên bảng màu và chỉ số bằng 0
rgb_pixels = npReshape[0,pixels]
print(rgb_pixels.shape)
#print(rgb_pixels)
# print(rgb_pixels[0:3])
# print(rgb_pixels[0, 1])
# Hiển thị ảnh từ mảng giá trị pixels
plt.imshow(rgb_pixels)
```

- 1 ảnh được tạo từ nhiều pixels
    - 1 pixels được tạo thành từ 3 màu RBG
